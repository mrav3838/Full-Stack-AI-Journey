```python
age = 20
is_student = True

if age < 5:
    print("You get free entry!")
elif age <= 17:
    print("You have to pay 8$")
elif age >= 18 and age <= 64:
    if is_student:
        print("You have to pay 9$")
    else:
        print("You have to pay 12$")
else:
    print("You have to pay 6$")
```
