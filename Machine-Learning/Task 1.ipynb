2. Practice all types of conversion

```python
#Types of Conversion
#1. Implicit Conversion
#2. Explicit Conversion
num_int = 10
num_flo = 5.5
result1 = num_int + num_flo # num_int is implicitly converted to float before addition
print(result1)

result2 = True + 4 # True is implicitly converted to 1 before addition
print(result2)

print(int(3.9))       # Output is turned into integer 3 from float 3.9
print(int("100"))     # Output is turned into integer 100 from string "100"

print(float(5))       # Output is turned into float 5.0 from integer 5
print(float("3.14"))  # Output is turned into float 3.14 from string "3.14"

print(complex(2, 3))  # Output is turned into complex number (2+3j)

print(str(42))        # Output is turned into string "42"

print(bool(0))        # Output is turned into boolean False
print(bool("Hello"))  # Output is turned into boolean True
```

3. Practice all types of operator with example

```python
a=8
b=2
print(a+b) #addition
print(a-b) #subtraction
print(a*b) #multiplication
print(a/b) #division
print(a//b)  # Floor division
print(a%b)   # Modulus
print(a**b)  # Exponentiation
print(a>b) #greater than
print(a<b) #less than
print(a==b) #equal to
print(a!=b) #not equal to
print(a>=b) #greater than or equal to
print(a<=b) #less than or equal to
print(a and b) #logical and
print(a or b) #logical or
print(not a) #logical not
print(a is b) #identity operator
print(a is not b) #identity operator
print(a in [1, 2, 3, 4, 5, 6, 7, 8]) #membership operator
print(a not in [1, 2, 3, 4, 5, 6, 7, 8])
```
